```python
n = np.array([1,4])
y = np.linspace(-1,1, 100)
v = np.zeros(len(y))

fig, ax = plt.subplots(1, 1, figsize = (5,5))

for i in range(len(n)):
    v = (1 - abs(y)**(n[i]+1))
    ax.plot(v,(0.5*y +0.5), label = f'n = {n[i]}')
    ax.legend()
    ax.set_title("cessation of subduction")
    # ax.set_xlim(0,1)
    # ax.set_ylim(0,1)
    ax.set_xlabel("v/v$_{max}$")
    ax.set_ylabel("y/h")
plt.savefig("plots/cessation.png", dpi = 500)
# plt.close()

```

```python
z = np.linspace(0,1, 100)
v = np.zeros(len(y))
E = np.array([1,3,10])

fig, ax = plt.subplots(1, 1, figsize = (5,5))

for i in range(len(E)):
    v = ((np.exp(E[i]) - np.exp(E[i]*z)))/(np.exp(E[i]) - 1)
    ax.plot(v,z, label = f'E = {E[i]}')
    ax.legend()
    ax.set_title("Teperature dependence, Newtonian")
    # ax.set_xlim(0,1)
    # ax.set_ylim(0,1)
    ax.set_xlabel("v'")
    ax.set_ylabel("z'")
plt.savefig("plots/Tdep_vel.png", dpi = 500)
plt.close()
```
